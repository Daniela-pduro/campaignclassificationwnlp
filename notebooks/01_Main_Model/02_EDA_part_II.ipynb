#### 3.2. Testing the variation in `total_sent`, `opens` and `clicks` caused by  `customer_cat` 
Steps:

- 3.2.1. Converting `customer_cat` into a categorical variable.

- 3.2.2. Performing one-way ANOVA for each continous variable.


#### 3.2.1. Converting  `customer_cat` into a categorical variable

```python
main_model_eda.customer_cat_to_cat_converter()
```

#### 3.2.2. Performing one-way ANOVA for each continous variable

- 3.2.2.1. Obtaining a random sample.
- 3.2.2.2. Check normality using QQ plots.
- 3.2.2.3. Check homogeneity variance.
- 3.2.2.4. Performing ANOVA.

#### 3.2.2.1. Obtaining a random sample

```python
campaign_sample, groups = main_model_eda.random_sampler(500,30)
```

```python
campaign_sample
```

```python
groups
```

#### 3.2.2.2. Checking normality using QQ plots

```python
main_model_eda.qq_plotter(campaign_sample, 'total_sent')
```

```python
main_model_eda.qq_plotter(campaign_sample, 'opens')
```

```python
main_model_eda.qq_plotter(campaign_sample, 'clicks')
```

- The points fall roughly on the straight line pattern meaning that the data comes from populations with a normal distribution.

#### 3.2.2.3. Checking homogeneity of variance

```python
ratio = main_model_eda.homog_var_checker(campaign_sample)
ratio
```

- The ratio of the largest to the smallest group is less than 1.5 meaning that we have equal group sizes.

#### 3.2.2.4. Performing ANOVA

- H<sub>0</sub>: `total_sent`, `opens` and `clicks` mean values are the same for every customer category.

- H<sub>1</sub>: `total_sent`, `opens` and `clicks` mean values are not the same for every customer category.

```python
total_sent_anova_table = main_model_eda.anova_tester(campaign_sample, 'total_sent', 0.05, "one_tailed")
total_sent_anova_table
```

```python
opens_anova_table = main_model_eda.anova_tester(campaign_sample, 'opens', 0.05, "one_tailed")
opens_anova_table
```

```python
clicks_anova_table = main_model_eda.anova_tester(campaign_sample, 'clicks', 0.05, "one_tailed")
clicks_anova_table
```
